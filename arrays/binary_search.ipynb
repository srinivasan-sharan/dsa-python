```python
def binary_search(arr: List[int], target: int) -> bool:
    left, right = 0, len(arr)-1
    while left <= right:
        mid = (left+right)//2
        if target == arr[mid]:
            return True
        if target < arr[mid]:
            right = mid - 1
        else:
            left = mid + 1

    return "element not found"
```

```python
binary_search(arr, 9952)
```
